## 3.1 Introduction to Vector Sets

In the previous chapter, we looked at vectors in isolation or in pairs (e.g., adding two vectors or taking their dot product). In this chapter, we zoom out to look at **sets of vectors** and how they interact.

When we deal with datasets in machine learning, we are rarely looking at just one or two vectors. A dataset with 1,000 features is a collection of 1,000 vectors. Understanding how these vectors relate to one another—whether they provide unique information or just repeat existing information—is the core theme of this chapter.

## 3.2 Linear Weighted Combinations

A **linear combination** is the most important operation involving a set of vectors. It simply means taking a set of vectors, multiplying each vector by a scalar (weight), and adding the results together.

Algebraically, if you have a set of vectors $v_1, v_2, \dots, v_n$ and a set of scalar weights $c_1, c_2, \dots, c_n$, the linear combination $w$ is:

$$ w = c_1v_1 + c_2v_2 + \dots + c_nv_n $$

In Data Science, a neural network layer computes exactly this: it takes input features (vectors), multiplies them by learned weights (scalars), and sums them up.

```python
# Define a set of three vectors in R^2
v1 = np.array([2, 5])
v2 = np.array([3, 1])
v3 = np.array([-2, 4])

# Define their corresponding scalar weights
c1, c2, c3 = 1.5, -2.0, 0.5

# Compute the linear weighted combination
w = (c1 * v1) + (c2 * v2) + (c3 * v3)

print(f"Vector 1: {v1}, Weight: {c1}")
print(f"Vector 2: {v2}, Weight: {c2}")
print(f"Vector 3: {v3}, Weight: {c3}\n")
print(f"Linear Combination Result (w): {w}")
```

## 3.3 Linear Independence vs. Linear Dependence

A set of vectors is **Linearly Dependent** if at least one vector in the set can be created by a linear combination of the other vectors. Geometrically, this means the dependent vector lies perfectly on the line, plane, or space created by the others—it adds no "new dimensions" or new information to the dataset.

A set of vectors is **Linearly Independent** if no vector can be constructed from the others. Every vector provides completely unique, non-redundant directional information.

Mathematical definition: A set is independent if the only way to satisfy the equation $c_1v_1 + c_2v_2 + \dots = 0$ is if all scalars $c_1, c_2, \dots$ are exactly zero.

```python
# Example of Linearly Dependent Vectors
d1 = np.array([1, 2, 3])
d2 = np.array([2, 4, 6]) # Exactly 2 * d1
d3 = np.array([4, 8, 12]) # Exactly 4 * d1 or 2 * d2

print("Linearly Dependent Set:")
print("d2 is just a scaled version of d1. It adds no new geometric dimension.")

# Example of Linearly Independent Vectors
i1 = np.array([1, 0, 0])
i2 = np.array([0, 1, 0])
i3 = np.array([0, 0, 1])

print("\nLinearly Independent Set:")
print("There is no way to multiply i1 and i2 by any scalars to create i3.")
```

## 3.4 Vector Spaces and Subspaces

A **Vector Space** is a collection of vectors that can be added together and multiplied by scalars to produce another vector that is *still inside* that same space. The most common vector spaces are $\mathbb{R}^2$ (the 2D XY plane) and $\mathbb{R}^3$ (3D space).

A **Subspace** is a smaller vector space nestled inside a larger one, subject to three strict rules:
1. It must contain the zero vector (the origin).
2. It must be closed under addition (adding any two vectors in the subspace keeps you in the subspace).
3. It must be closed under scalar multiplication.

For example, a 2D plane passing through the origin `[0,0,0]` inside a 3D space is a valid subspace of $\mathbb{R}^3$.

## 3.5 The Concept of 'Span'

The **Span** of a set of vectors is the continuous geometric space created by taking **all possible linear combinations** of those vectors (using every real number from $-\infty$ to $+\infty$ as weights).

- The span of a single non-zero vector in 2D is an infinitely long 1D line passing through the origin.
- The span of two linearly independent vectors in 3D is an infinitely wide 2D plane cutting through the origin.
- The span of two linearly dependent vectors in 3D is just a 1D line, because the second vector fails to provide a new direction to pull the space into a plane.

Let's visualize the span of two vectors in a 3D space.

```python
# Define two linearly independent vectors in R^3
v1 = np.array([1, 2, 0])
v2 = np.array([0, 1, 1])

# To visualize their span (a 2D plane in 3D space), we will create many random linear combinations
num_points = 500
span_points = []

for _ in range(num_points):
    c1 = np.random.uniform(-5, 5)
    c2 = np.random.uniform(-5, 5)
    point = (c1 * v1) + (c2 * v2)
    span_points.append(point)

span_points = np.array(span_points)

# Plotting the 2D plane (span) in 3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter the linear combinations to show the surface of the plane
ax.scatter(span_points[:, 0], span_points[:, 1], span_points[:, 2], c='c', alpha=0.3, s=5)

# Plot the original vectors that create this span
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', linewidth=3, arrow_length_ratio=0.1, label='v1')
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='b', linewidth=3, arrow_length_ratio=0.1, label='v2')

ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Span of two independent vectors in R^3 (Forms a 2D Plane)')
plt.legend()
plt.show()
```

## 3.6 Basis and Orthogonal Basis

A **Basis** is a set of vectors that acts as a reference frame (or ruler) for a vector space. For a set of vectors to be a basis for a space, it must satisfy two conditions:
1. They must **span** the entire space (e.g., you need exactly 3 vectors to span $\mathbb{R}^3$).
2. They must be **linearly independent** (no redundant vectors).

The standard basis for our everyday Cartesian coordinate system consists of vectors that are length 1 and point strictly along the X, Y, and Z axes. Because these vectors are at exactly 90-degree angles to one another, they form an **Orthogonal Basis**. If their lengths are all exactly 1, they form an **Orthonormal Basis**.

Let's look at the standard basis in Python, which is identical to the Identity Matrix.

```python
# The standard basis for R^3
e1 = np.array([1, 0, 0])
e2 = np.array([0, 1, 0])
e3 = np.array([0, 0, 1])

# Verify Orthogonality (Dot product of any pair must be 0)
print(f"Dot product e1 . e2: {np.dot(e1, e2)}")
print(f"Dot product e2 . e3: {np.dot(e2, e3)}")
print(f"Dot product e1 . e3: {np.dot(e1, e3)}\n")

# Verify Normality (Length of each must be 1)
print(f"Length of e1: {np.linalg.norm(e1)}")
print(f"Length of e2: {np.linalg.norm(e2)}")
print(f"Length of e3: {np.linalg.norm(e3)}\n")

print("Since they are orthogonal and have a length of 1, they form an ORTHONORMAL basis.")
```

To summarize Chapter 3:

- **Linear Independence** is the concept of measuring unique information. If a feature in your dataset is just a linear combination of other features, it is dependent and mathematically redundant.
- **Span** is the reachable space generated by a set of vectors.
