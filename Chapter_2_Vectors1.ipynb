## 2.1 The Two Perspectives of a Vector

Vectors are the most fundamental building blocks of linear algebra. In data science, you can think of vectors through two distinct but complementary lenses:

1. **The Algebraic (Computational) Perspective:** A vector is simply an ordered list of numbers. Each number represents a coordinate in a specific dimension or a specific feature of a dataset. For example, a vector containing `[Age, Weight, Height]` is a 3-dimensional vector describing a person.
2. **The Geometric (Visual) Perspective:** A vector is a directed line segment (an arrow) in space. It has a specific **length (magnitude)** and points in a specific **direction**. The arrow traditionally starts at the origin of the coordinate system `[0, 0]` and its tip ends at the coordinates specified by the vector's numbers.

In this chapter, we will explore vector arithmetic and see how algebraic operations affect the geometric arrows in a 2D space.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set plotting style for better visibility
%matplotlib inline
print("Scientific libraries successfully imported.")
```

## 2.2 Vector Addition and Subtraction

**Algebraic Addition:**
To add two vectors, they **must have the same dimensionality** (the same number of elements). You simply add the corresponding elements together. If $v = [1, 2]$ and $w = [3, 4]$, then $v + w = [1+3, 2+4] = [4, 6]$.

**Geometric Addition (Head-to-Tail Method):**
To visualize $v + w$, draw vector $v$ starting from the origin. Then, draw vector $w$, but instead of starting it at the origin, start it at the tip (head) of vector $v$. The resulting vector is drawn from the origin to the tip of $w$.

Let us visualize this in Python.

```python
v1 = np.array([3, 1])
v2 = np.array([1, 4])

# Algebraic addition
v3 = v1 + v2
print(f"v1: {v1}")
print(f"v2: {v2}")
print(f"Result of v1 + v2: {v3}\n")

# Geometric Visualization
plt.figure(figsize=(6, 6))

# Plot v1
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
# Plot v2 starting from the tip of v1
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', label='v2 (shifted)')
# Plot the resultant vector v3
plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='red', label='v1 + v2')

plt.xlim(-1, 5)
plt.ylim(-1, 6)
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.grid(alpha=0.4)
plt.legend()
plt.title('Vector Addition: Head-to-Tail Method')
plt.show()
```

## 2.3 Scalar-Vector Multiplication

A scalar is a single number (like $2$, $-0.5$, or $0$). When you multiply a vector by a scalar, you multiply every element of the vector by that number.

Geometrically, a scalar changes the **scale** (length) of the vector but not its fundamental line of direction. 
- If the scalar is $> 1$, the vector stretches.
- If the scalar is between $0$ and $1$, the vector shrinks.
- If the scalar is $< 0$ (negative), the vector flips $180^{\circ}$ to point in the exact opposite direction.

```python
v = np.array([2, 1.5])
s1 = 2      # Stretches the vector
s2 = -0.5   # Shrinks and flips the vector

plt.figure(figsize=(6, 6))

# Using plt.plot for a different visualization style
plt.plot([0, v[0]], [0, v[1]], 'b', label='Original v', linewidth=3)
plt.plot([0, s1*v[0]], [0, s1*v[1]], 'r--', label='2 * v (Stretched)', linewidth=2)
plt.plot([0, s2*v[0]], [0, s2*v[1]], 'g-.', label='-0.5 * v (Flipped & Shrunk)', linewidth=2)

plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.grid(alpha=0.4)
plt.legend()
plt.title('Scalar-Vector Multiplication')
plt.show()
```

## 2.4 Vector-Vector Multiplication: The Dot Product

The **Dot Product** (also known as the inner product) is arguably the most important operation in all of linear algebra and machine learning. 

The dot product takes two vectors of the same length and returns a **single scalar number**.
Algebraically, it is calculated by multiplying corresponding elements and summing all the products together:

$$ a \cdot b = a^T b = \sum_{i=1}^{n} a_i b_i = a_1 b_1 + a_2 b_2 + \dots + a_n b_n $$

The dot product is the backbone of measuring similarity, computing correlations, filtering signals, and projecting data.

```python
v1 = np.array([1, 2, 3, 4])
v2 = np.array([0, -1, 2, -2])

# There are multiple ways to compute the dot product in NumPy:

# 1. Using np.dot() function
dot_1 = np.dot(v1, v2)

# 2. Using the @ operator (Python 3.5+)
dot_2 = v1 @ v2

# 3. Element-wise multiplication followed by summation (Algebraic proof)
dot_3 = np.sum(v1 * v2)

print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}\n")
print(f"Result using np.dot : {dot_1}")
print(f"Result using @      : {dot_2}")
print(f"Result using sum    : {dot_3}")
```

## 2.5 Hadamard Multiplication (Element-wise)

Do not confuse the dot product with the **Hadamard Product**. The Hadamard product simply multiplies corresponding elements but **returns a new vector**, not a scalar. This is very common in programming (using the `*` operator) but less common in pure linear algebra theory unless specifically indicated (often denoted by $\odot$).

```python
v1 = np.array([5, 4, 8, 2])
v2 = np.array([1, 0, 0.5, -1])

hadamard = v1 * v2

print("Vector 1         :", v1)
print("Vector 2         :", v2)
print("Hadamard Product :", hadamard, "<-- Notice this is an array, not a single number")
```

## 2.6 Vector Magnitude (Length) and Unit Vectors

The **magnitude** (or length) of a vector is the distance from its tail at the origin to its head. This is also called the **L2 Norm** or Euclidean norm, denoted as $||v||$. It is calculated using the Pythagorean theorem:

$$ ||v|| = \sqrt{v_1^2 + v_2^2 + \dots + v_n^2} $$

A **Unit Vector** is a vector that has a magnitude of exactly $1$. It represents pure direction without being scaled by distance. You create a unit vector by dividing a vector by its own magnitude: $u = \frac{v}{||v||}$.

```python
# Consider a 3-4-5 right triangle vector
v = np.array([3, 4])

# Calculate magnitude using numpy's linear algebra module
v_mag = np.linalg.norm(v)
print(f"Original vector: {v}")
print(f"Magnitude of the vector: {v_mag}")

# Create a unit vector
v_unit = v / v_mag
print(f"\nUnit vector: {v_unit}")
print(f"Verify the magnitude of the unit vector is exactly 1: {np.linalg.norm(v_unit)}")
```

## 2.7 The Geometric Perspective of the Dot Product

The dot product also has a powerful geometric definition involving the angle $\theta$ between two vectors:

$$ a \cdot b = ||a|| \, ||b|| \cos(\theta) $$

This equation means the dot product is a measure of how much two vectors "overlap" or point in the same direction. We can rearrange this formula to compute the exact angle between any two vectors in any N-dimensional space:

$$ \theta = \arccos\left( \frac{a \cdot b}{||a|| \, ||b||} \right) $$

```python
def calculate_angle(v1, v2):
    # 1. Compute dot product
    dot_prod = np.dot(v1, v2)
    
    # 2. Compute magnitudes
    mag_v1 = np.linalg.norm(v1)
    mag_v2 = np.linalg.norm(v2)
    
    # 3. Calculate cosine of the angle
    cos_theta = dot_prod / (mag_v1 * mag_v2)
    
    # 4. Use arc-cosine to get the angle in radians, then convert to degrees
    angle_rad = np.arccos(cos_theta)
    angle_deg = np.rad2deg(angle_rad)
    
    return angle_deg

vector_a = np.array([1, 0])
vector_b = np.array([1, 1])

print(f"The angle between {vector_a} and {vector_b} is {calculate_angle(vector_a, vector_b):.2f} degrees.")
```

## 2.8 Orthogonal Vectors

Two vectors are called **orthogonal** if they meet at exactly a $90^{\circ}$ angle. 

Looking back at the cosine rule: since $\cos(90^{\circ}) = 0$, **the dot product of two orthogonal vectors is always exactly 0**. This property is essential in algorithms like PCA, where we aim to extract uncorrelated (orthogonal) features from a dataset.

```python
# Define two orthogonal vectors
v1 = np.array([2, 2])
v2 = np.array([-2, 2])

dp = np.dot(v1, v2)
print(f"Dot product of {v1} and {v2} = {dp}")

plt.figure(figsize=(5, 5))
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')

plt.xlim(-3, 3)
plt.ylim(0, 4)
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.grid(alpha=0.4)
plt.legend()
plt.title(f'Orthogonal Vectors (Dot Product = {dp})')
plt.show()
```
