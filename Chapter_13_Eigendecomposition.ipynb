## 13.1 Introduction: The Core Engine of a Matrix

When we multiply a vector by a matrix ($Ax$), the matrix acts as a transformation that typically rotates and stretches the vector into a completely new direction.

However, for almost every square matrix, there exists a special set of vectors that **do not rotate** when transformed by the matrix. They only get stretched or shrunk (or flipped 180 degrees). These special vectors are called **Eigenvectors**.

The amount by which an eigenvector gets stretched or shrunk is a single scalar number called an **Eigenvalue**.

Mathematically, this is expressed in the most famous equation in linear algebra:
$$ \mathbf{A}\mathbf{v} = \lambda\mathbf{v} $$

- $\mathbf{A}$ is a square matrix.
- $\mathbf{v}$ is the Eigenvector.
- $\lambda$ (lambda) is the Eigenvalue (a scalar).

This equation elegantly states that transforming the eigenvector by the matrix $\mathbf{A}$ yields the exact same result as simply scaling it by $\lambda$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set plotting style
%matplotlib inline
np.random.seed(42)
print("Scientific libraries successfully imported for Chapter 13.")
```

## 13.2 Geometric Intuition of Eigenvectors

Let's visualize the difference between an ordinary vector and an eigenvector when both are multiplied by the same transformation matrix.

- The ordinary vector will be knocked off its original span (it will rotate).
- The eigenvector will stay perfectly on its original span, only changing its length.

```python
# Define a 2x2 Transformation Matrix
A = np.array([
    [1, 2],
    [2, 1]
])

# A known eigenvector of this matrix (we will learn how to find this shortly)
v_eigen = np.array([1, 1])

# A random ordinary vector
v_random = np.array([1, 0])

# Apply the transformation
Av_eigen = A @ v_eigen
Av_random = A @ v_random

# Visualization
plt.figure(figsize=(10, 5))

# Plot 1: The Ordinary Vector
plt.subplot(1, 2, 1)
plt.plot([0, v_random[0]], [0, v_random[1]], 'k-', linewidth=3, label='Original Vector')
plt.plot([0, Av_random[0]], [0, Av_random[1]], 'r--', linewidth=3, label='Transformed (Rotated!)')
plt.title('Ordinary Vector')
plt.xlim(0, 3); plt.ylim(0, 3)
plt.grid(alpha=0.3); plt.legend()

# Plot 2: The Eigenvector
plt.subplot(1, 2, 2)
plt.plot([0, v_eigen[0]], [0, v_eigen[1]], 'k-', linewidth=3, label='Original Eigenvector')
plt.plot([0, Av_eigen[0]], [0, Av_eigen[1]], 'g--', linewidth=3, label='Transformed (Scaled only!)')
plt.title('Eigenvector')
plt.xlim(0, 4); plt.ylim(0, 4)
plt.grid(alpha=0.3); plt.legend()

plt.tight_layout()
plt.show()

print("Notice how the transformed eigenvector stays on the exact same line as the original vector.")
```

## 13.3 Finding Eigenvalues and Eigenvectors in Python

Calculating eigenvalues by hand involves finding the roots of the characteristic polynomial, $\det(\mathbf{A} - \lambda\mathbf{I}) = 0$. For matrices larger than 3x3, this is practically impossible to do manually.

In Python, we use `np.linalg.eig()`, which returns two outputs:
1. An array of **eigenvalues**.
2. A matrix where every **column** is a corresponding normalized eigenvector.

```python
# Create a 3x3 matrix
M = np.array([
    [4, 2, 0],
    [2, 4, 2],
    [0, 2, 4]
])

# Compute Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(M)

print("Eigenvalues (lambda):")
print(np.round(eigenvalues, 3))

print("\nEigenvectors Matrix (Columns are eigenvectors):")
print(np.round(eigenvectors, 3))

# Verify the equation A*v = lambda*v for the first eigenpair
lambda_1 = eigenvalues[0]
v_1 = eigenvectors[:, 0]

left_side = M @ v_1
right_side = lambda_1 * v_1

print("\nVerification (A * v):", np.round(left_side, 3))
print("Verification (lambda * v):", np.round(right_side, 3))
print("They match perfectly!")
```

## 13.4 Diagonalization (The Eigendecomposition Formula)

If we take all the eigenvectors and pack them into a matrix $\mathbf{V}$, and take all the eigenvalues and put them on the diagonal of a matrix $\mathbf{\Lambda}$ (Lambda), we can rewrite the equation $\mathbf{A}\mathbf{v} = \lambda\mathbf{v}$ for all vectors simultaneously:

$$ \mathbf{A}\mathbf{V} = \mathbf{V}\mathbf{\Lambda} $$

If we isolate $\mathbf{A}$ by multiplying both sides by $\mathbf{V}^{-1}$, we get the **Eigendecomposition** of the matrix:
$$ \mathbf{A} = \mathbf{V}\mathbf{\Lambda}\mathbf{V}^{-1} $$

This allows us to completely reconstruct the original matrix using only its fundamental "DNA" (its eigenvalues and eigenvectors).

```python
# Matrix V contains the eigenvectors
V = eigenvectors

# Matrix Lambda is a diagonal matrix containing the eigenvalues
Lambda = np.diag(eigenvalues)

# Compute V^{-1}
V_inv = np.linalg.inv(V)

# Reconstruct A using V * Lambda * V^-1
M_reconstructed = V @ Lambda @ V_inv

print("Original Matrix M:")
print(M)

print("\nReconstructed Matrix (V @ Lambda @ V^-1):")
print(np.round(M_reconstructed))
```

## 13.5 The Magic of Symmetric Matrices (Spectral Theorem)

Not all matrices have real eigenvalues. Sometimes, trying to find roots results in complex numbers (involving $i$, the imaginary unit). Also, matrix $\mathbf{V}$ is not always invertible.

However, there is a special rule in linear algebra called the **Spectral Theorem**. It states that if a matrix is **Symmetric** (meaning $\mathbf{A} = \mathbf{A}^T$), two magical things are guaranteed to happen:
1. All of its eigenvalues will be **real numbers** (no complex/imaginary numbers).
2. All of its eigenvectors will be **perfectly orthogonal** to each other ($90^{\circ}$ apart).

Because the eigenvector matrix $\mathbf{V}$ is orthogonal, its inverse is simply its transpose ($\mathbf{V}^{-1} = \mathbf{V}^T$). For symmetric matrices, the decomposition becomes incredibly fast and stable:
$$ \mathbf{A} = \mathbf{V}\mathbf{\Lambda}\mathbf{V}^T $$

Because Covariance matrices (from Chapter 7) are always symmetric, this theorem is the direct mathematical foundation of Principal Component Analysis (PCA)!

```python
# 1. Create a random symmetric matrix by computing A^T * A
A_random = np.random.randn(4, 4)
Symmetric_Matrix = A_random.T @ A_random

# For symmetric matrices, we use np.linalg.eigh() instead of eig().
# The 'h' stands for Hermitian (the complex equivalent of symmetric).
# It uses a faster, more numerically stable algorithm.
evals_sym, evecs_sym = np.linalg.eigh(Symmetric_Matrix)

print("Eigenvalues of the Symmetric Matrix (All real numbers!):")
print(np.round(evals_sym, 3))

# 2. Verify that the Eigenvectors form an Orthogonal Matrix
# If V is orthogonal, then V^T @ V must equal the Identity Matrix.
orthogonality_check = evecs_sym.T @ evecs_sym

print("\nVerify V^T @ V for symmetric eigenvectors:")
print(np.round(orthogonality_check, 3))
print("It is the perfect Identity Matrix! The eigenvectors are orthogonal.")

# 3. Reconstruct using V @ Lambda @ V^T (No inverse needed!)
Lambda_sym = np.diag(evals_sym)
Reconstructed_Sym = evecs_sym @ Lambda_sym @ evecs_sym.T

print("\nIs the reconstruction accurate using only the transpose?", 
      np.allclose(Symmetric_Matrix, Reconstructed_Sym))
```

## 13.6 Chapter Summary

In Chapter 13, we uncovered the fundamental DNA of matrices:
- **Eigenvectors** are the axes of transformation that do not rotate; they are only scaled.
- **Eigenvalues** are the scalar amounts by which the eigenvectors are stretched or shrunk.
- **Eigendecomposition** factors a matrix into its eigenvectors and eigenvalues: $\mathbf{A} = \mathbf{V}\mathbf{\Lambda}\mathbf{V}^{-1}$.
- **Symmetric Matrices** are the most well-behaved matrices in linear algebra, guaranteeing real eigenvalues and perfectly orthogonal eigenvectors, allowing for the much faster decomposition $\mathbf{A} = \mathbf{V}\mathbf{\Lambda}\mathbf{V}^T$.
