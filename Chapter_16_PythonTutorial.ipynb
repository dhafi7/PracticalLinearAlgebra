## 16.2 Standard Python Lists vs. NumPy Arrays

The most common mistake beginners make is confusing a standard Python `list` with a `NumPy array`. 

- A **Python List** is a general-purpose container. It can hold mixed data types (numbers, strings, other lists) and is optimized for dynamic resizing, but it does NOT support mathematical vector operations.
- A **NumPy Array** (`ndarray`) is a strictly typed, memory-efficient grid of numbers written in C. It supports element-wise math, linear algebra operations, and broadcasting.

Let's observe what happens when we try to "multiply" them.

```python
# 1. Python List
py_list = [1, 2, 3]

# 2. NumPy Array
np_array = np.array([1, 2, 3])

print("--- Scalar Multiplication ( * 3 ) ---")
# Multiplying a list duplicates its contents (Concatenation)
print("Python List * 3  :", py_list * 3)

# Multiplying a NumPy array scales its values mathematically
print("NumPy Array * 3  :", np_array * 3)

print("\n--- Addition ( + ) ---")
list_2 = [4, 5, 6]
array_2 = np.array([4, 5, 6])

# Adding lists joins them together
print("List + List      :", py_list + list_2)

# Adding arrays performs element-wise vector addition
print("Array + Array    :", np_array + array_2)
```

## 16.3 Array Creation and Dimensionality

NumPy provides several functions to generate matrices quickly without having to type them out manually. This is crucial for initializing weights in neural networks or creating dummy datasets.

Always check the `.shape` attribute to verify the dimensionality of your data!

```python
# np.zeros: Creates an array filled with 0s (Useful for memory pre-allocation)
Z = np.zeros((2, 4))
print("Zeros Matrix (2x4):\n", Z)

# np.ones: Creates an array filled with 1s
O = np.ones((3, 2))
print("\nOnes Matrix (3x2):\n", O)

# np.arange: Creates a sequence of numbers (start, stop_exclusive, step)
seq = np.arange(0, 10, 2)
print("\nSequence (arange):", seq)

# np.linspace: Creates evenly spaced numbers over a specified interval (Crucial for plotting smooth math functions)
lin = np.linspace(0, 1, 5)
print("\nLinspace (5 points between 0 and 1):", lin)

# np.random.randn: Standard normal distribution (mean 0, variance 1)
R = np.random.randn(3, 3)
print("\nRandom Matrix (Gaussian):\n", np.round(R, 2))
```

## 16.4 Matrix Indexing and Slicing

Extracting specific rows, columns, or sub-matrices is a daily task in Data Science. 
NumPy uses zero-based indexing. For a 2D matrix, the syntax is `Matrix[row_index, column_index]`.

The colon `:` is the slice operator. By itself, it means "give me everything in this dimension".

```python
# Create a 4x4 matrix counting from 1 to 16
M = np.arange(1, 17).reshape(4, 4)
print("Matrix M:\n", M)

print("\n--- Slicing Examples ---")
# 1. Get a specific element (Row 2, Column 1) -> Note: Indexing starts at 0!
print("Element at index [2, 1] :", M[2, 1])

# 2. Get an entire row (Row 0)
print("Entire Row 0            :", M[0, :])

# 3. Get an entire column (Column 3)
print("Entire Column 3         :", M[:, 3])

# 4. Sub-matrix extraction (Rows 1 and 2, Columns 1 and 2)
sub_matrix = M[1:3, 1:3]
print("\nSub-matrix (Rows 1-2, Cols 1-2):\n", sub_matrix)
```

## 16.5 Plotting with Matplotlib

Visualizing vectors, matrices, and data is arguably as important as calculating them. `matplotlib.pyplot` operates much like a canvas. You add elements to it sequentially, and then call `plt.show()` to render it.

When working with multiple distinct visualizations, using `plt.subplots()` is the cleanest, most professional way to organize your figures.

```python
# Generate some data
x = np.linspace(-5, 5, 100)
y_linear = x
y_quadratic = x**2
y_cubic = x**3

# Create a figure with 1 row and 3 columns of subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 4))

# Plot 1: Linear
axes[0].plot(x, y_linear, color='blue', linewidth=2)
axes[0].set_title("Linear (y = x)")
axes[0].grid(True)
axes[0].axhline(0, color='black', linewidth=1)
axes[0].axvline(0, color='black', linewidth=1)

# Plot 2: Quadratic
axes[1].plot(x, y_quadratic, color='red', linestyle='--')
axes[1].set_title("Quadratic (y = x^2)")
axes[1].grid(True)

# Plot 3: Cubic
axes[2].plot(x, y_cubic, color='green', linestyle='-.')
axes[2].set_title("Cubic (y = x^3)")
axes[2].grid(True)

# Optimize spacing between subplots
plt.tight_layout()
plt.show()
```
