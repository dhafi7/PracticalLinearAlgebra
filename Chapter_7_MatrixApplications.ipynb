## 7.2 Application 1: Geometric Transformations (Rotation Matrix)

In Chapter 6, we saw a brief example of a "Shear" matrix. Another highly common transformation is **Rotation**. 

To rotate a 2D vector (or a set of vectors) by a specific angle $\theta$ (in radians), we multiply it by the standard 2D Rotation Matrix:

$$ R = \begin{bmatrix} \cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) \end{bmatrix} $$

By multiplying our data matrix by $R$, every single point is instantly rotated around the origin. This technique is widely used in Computer Vision to augment image datasets (creating rotated copies of images so neural networks learn rotational invariance).

```python
# 1. Define a shape (a simple house) using 2D coordinates
# Each column is a [x, y] coordinate.
house = np.array([
    [0, 2, 2, 1, 0, 0],  # X coordinates
    [0, 0, 2, 3, 2, 0]   # Y coordinates
])

# 2. Define the Rotation Matrix (Let's rotate by 90 degrees = pi/2 radians)
theta = np.pi / 2
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

# 3. Apply the transformation (R @ house)
rotated_house = R @ house

# 4. Visualization
plt.figure(figsize=(8, 8))
plt.plot(house[0, :], house[1, :], 'b-o', linewidth=2, markersize=8, label='Original')
plt.plot(rotated_house[0, :], rotated_house[1, :], 'r-s', linewidth=2, markersize=8, label='Rotated 90 Deg')

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.grid(alpha=0.3)
plt.legend()
plt.title('Geometric Transformation: Rotation Matrix')
plt.gca().set_aspect('equal') # Ensure the grid is perfectly square
plt.show()
```

## 7.3 Application 2: The Covariance Matrix

In statistics, variance measures how much a single variable spreads out, while covariance measures how two variables move together. If you have a dataset with $N$ features, calculating the covariance between every possible pair of features one by one requires tedious loops.

Linear Algebra offers a breathtakingly elegant solution. If $\mathbf{X}$ is a mean-centered data matrix (where columns are features and rows are observations), the entire **Covariance Matrix** $\mathbf{C}$ can be computed with a single matrix multiplication:

$$ \mathbf{C} = \frac{1}{n-1} \mathbf{X}^T \mathbf{X} $$

This resulting matrix $\mathbf{C}$ is a symmetric square matrix. The diagonal elements are the variances of each feature, and the off-diagonal elements are the covariances between features. This matrix is the starting point for Principal Component Analysis (PCA).

```python
# 1. Simulate a dataset with 3 features and 100 observations (100x3 matrix)
n_samples = 100
f1 = np.random.randn(n_samples) * 2
f2 = f1 * 1.5 + np.random.randn(n_samples) # Strongly correlated with f1
f3 = np.random.randn(n_samples) * 5        # Independent feature

X = np.column_stack((f1, f2, f3))
print(f"Data Matrix X shape: {X.shape}")

# 2. Mean-center the data (Subtract the mean of each column)
X_centered = X - np.mean(X, axis=0)

# 3. Compute the Covariance Matrix using Pure Linear Algebra (X^T @ X)
# Note: We divide by (n-1) for sample covariance
cov_matrix_manual = (X_centered.T @ X_centered) / (n_samples - 1)

# 4. Compare with NumPy's built-in statistical function
# Note: np.cov expects features as rows, so we pass X.T
cov_matrix_numpy = np.cov(X.T)

print("\nCovariance Matrix (Computed via Linear Algebra X^T @ X):")
print(np.round(cov_matrix_manual, 2))

print("\nCovariance Matrix (Computed via np.cov):")
print(np.round(cov_matrix_numpy, 2))

print("\nNotice how they match perfectly! Linear algebra is the engine of statistics.")
```

## 7.4 Application 3: Image Processing as Matrix Manipulation

A grayscale image is literally just a 2D matrix where every element represents the brightness of a pixel (usually from 0 for black to 255 for white). A color image is simply three of these matrices stacked together (Red, Green, Blue).

Because images are matrices, any matrix operation (like transposition, scalar multiplication, or addition) directly translates into a visual effect on the image.

```python
# 1. Create a synthetic "Image" matrix (using a mathematical function to make it look interesting)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X_grid, Y_grid = np.meshgrid(x, y)

# Create ripples using a sine wave function based on distance from center
Z_image = np.sin(np.sqrt(X_grid**2 + Y_grid**2))

# 2. Perform Matrix Operations on the Image
# Transposing the matrix mirrors the image across its diagonal
Z_transposed = Z_image.T 

# Inverting the colors (Assuming normalized values between -1 and 1, multiplying by -1 inverts it)
Z_inverted = Z_image * -1 

# 3. Visualizing the matrices as images using plt.imshow
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(Z_image, cmap='gray')
axes[0].set_title('Original Matrix (Image)')
axes[0].axis('off')

axes[1].imshow(Z_transposed, cmap='gray')
axes[1].set_title('Transposed Matrix (Mirrored)')
axes[1].axis('off')

axes[2].imshow(Z_inverted, cmap='gray')
axes[2].set_title('Scalar Mult * -1 (Color Inverted)')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
