## 1.1 Motivation and Philosophy

Linear algebra is the foundational mathematical language of modern computational sciences, including Data Science, Machine Learning, and Deep Learning. Virtually all data—whether text tokens, tabular sheets, financial time series, or pixel arrays—must be transformed into numerical vectors and matrices before algorithms can extract patterns from them.

This book approaches linear algebra with a specific philosophy designed for practitioners:
- **The Dual Intuition:** Developing both **Algebraic Intuition** (the numerical and algorithmic mechanics) and **Geometric Intuition** (visualizing vectors as points or directional arrows in space, and matrices as operations that transform that space).
- **Code-Centric Learning:** Bridging the gap between rigid mathematical proofs and concrete programmatic implementations using Python and its underlying numerical libraries.

## 1.2 Applications in Data Science

Linear algebra does not exist in a vacuum; it is the engine powering operations such as:
1. **Dimensionality Reduction:** Techniques like Principal Component Analysis (PCA) rely on eigendecomposition and Singular Value Decomposition (SVD) to project high-dimensional feature spaces onto low-dimensional planes.
2. **Image and Signal Processing:** Digital images are stored as matrices where transformations like rotation, filtering, or compression are executed via matrix operations.
3. **Model Optimization:** Linear regression, neural network layer weights, and optimization algorithms (like gradient descent) utilize vector dot products and matrix multiplications to compute predictions and update parameters efficiently.

## 1.4 Mathematical vs. Programming Conventions

One major source of confusion for beginners is the slight discrepancy between traditional mathematical textbook notations and programming paradigms:

| Concept | Mathematical Standard | Python / NumPy Convention |
| :--- | :--- | :--- |
| **Indexing** | 1-based indexing ($v_1, v_2, \dots$ ) | 0-based indexing (`v[0], v[1], ...`) |
| **Vectors** | Assumed to be **Column Vectors** by default | 1D arrays are orientationless; explicitly 2D arrays are needed for orientation |
| **Matrices** | Capital letters ($A, B, X$) with size $m \times n$ | 2D arrays with explicit row and column shapes |

Let us explore these properties programmatically.

## 1.5 Building Vectors in NumPy

In programming, a simple list of numbers is a 1D array. In math, a vector must have a clear direction: either a **Row Vector** ($1 \times n$) or a **Column Vector** ($m \times 1$). We can inspect these dimensional boundaries using the `.shape` attribute.

```python
# Creating a basic 1D array (Orientationless vector)
v_1d = np.array([1, 2, 3])
print("1D Array Vector:", v_1d)
print("Shape of 1D Array:", v_1d.shape)
print("Notice that it only has one dimension value, meaning it has no explicit row or column orientation.\n")

# Creating an explicit Row Vector (using double brackets)
v_row = np.array([[1, 2, 3]])
print("Explicit Row Vector:\n", v_row)
print("Shape of Row Vector:", v_row.shape, "-> (1 row, 3 columns)\n")

# Creating an explicit Column Vector
v_col = np.array([[1],
                  [2],
                  [3]])
print("Explicit Column Vector:\n", v_col)
print("Shape of Column Vector:", v_col.shape, "-> (3 rows, 1 column)")
```

## 1.6 Vector Transposition

Transposition is the operation that flips an oriented vector or matrix over its diagonal, turning rows into columns and columns into rows. In mathematics, this is indicated by a superscript $v^T$. In NumPy, we use the `.T` attribute.

*Crucial Note: Transposing a standard 1D array (`shape=(n,)`) does absolutely nothing because it lacks a second dimension axis. It must be explicitly 2D for transposition to take visual effect.*

```python
print("Transposing a 1D array directly:", v_1d.T, "Shape:", v_1d.T.shape)
print("Result: No change because a 1D array has no second axis to flip.\n")

# Transposing an explicit row vector into a column vector
row_to_col = v_row.T
print("Transposed Row Vector (Now Column):\n", row_to_col)
print("New Shape:", row_to_col.shape)

# Alternative: Reshaping a 1D array into an oriented vector
v_reshaped_col = v_1d.reshape(-1, 1)
print("\nReshaped from 1D to Column Vector:\n", v_reshaped_col)
print("Shape after reshape:", v_reshaped_col.shape)
```

## 1.7 Geometric Representation of Vectors

Geometrically, a vector represents an instruction for moving through space, originating from a starting reference point (usually the origin $[0,0]$). For example, the vector $v = [2, 3]$ tells us to move 2 units along the X-axis and 3 units along the Y-axis.

Let us create a plot using `plt.quiver` to visualize this core concept.

```python
# Define the vector components [X_direction, Y_direction]
v = np.array([2, 3])

plt.figure(figsize=(6, 6))

# Plot the vector starting from the origin (0,0)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector v = [2, 3]')

# Adjust the graph grid parameters
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.axhline(0, color='black', linewidth=1.2)
plt.axvline(0, color='black', linewidth=1.2)
plt.grid(True, linestyle='--', alpha=0.6)
plt.title('Geometric Intuition: Vector as a Directed Arrow')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.legend()
plt.show()
```
